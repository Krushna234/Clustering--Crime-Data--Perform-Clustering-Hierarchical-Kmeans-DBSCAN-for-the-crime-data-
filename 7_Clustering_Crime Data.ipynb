#### Perform clustering (hierarchical,K means clustering and DBSCAN) for the Crime data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from scipy.cluster import hierarchy
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
```

# Hierarchical

```python
df = pd.read_csv("C:\\Users\\krush\\Desktop\\Data Science\\Assignments\\Clustering\\crime_data.csv", encoding='latin-1')
df
```

## EDA and Visualization

```python
df.rename(columns={'Unnamed: 0':'State' }, inplace= True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

## Scaling down the data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x = pd.DataFrame(sc.fit_transform(df.iloc[:,1: ]), columns=df.iloc[:,1: ].columns)
x
```

## Finding optimal number of clusters using dendograms

```python
from scipy.cluster import hierarchy
lk = hierarchy.linkage(x, method='complete')
```

```python
dendrogram = hierarchy.dendrogram(lk)
plt.figure(figsize = (25,5))
plt.title('Hierarchical Dendrogram')
plt.show()
```

## Model Building

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=4)
```

```python
ylabel = hc.fit_predict(x)
ylabel
```

```python
df['Cluster'] = ylabel
df
```

```python
df.groupby('Cluster').agg('mean')
```

## Visualizing the clusters

```python
df[df['Cluster']==3] 
```

```python
sns.scatterplot(df['Assault'], df['Cluster'], hue = df['State'])
plt.legend(bbox_to_anchor = (1,1))
```

# Kmeans

## Basic Model

```python
from sklearn.cluster import KMeans
km = KMeans()
```

```python
cluster2 = km.fit_predict(x)
cluster2
```

## Finding optimal number of clusters using Elbow Graph

```python
wcss2 = []

for i in range (1,20):
    km = KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss2.append(km.inertia_)
```

```python
wcss2
```

```python
plt.plot(range(1,20), wcss2, color = 'Red')
plt.scatter(range(1,20), wcss2, color = 'Blue')
plt.title('ELBOW METHOD')
plt.xlabel('Kvalues')
plt.ylabel('WCSS')
plt.show()
```

## Revised Model

```python
# k-value = 6
km = KMeans(n_clusters=4)
ylabel2 = km.fit_predict(x)
ylabel2
```

```python
df['Cluster2'] = ylabel2
df.head()
```

## Visualizing the clusters

```python
df[df['Cluster2']==3]
```

```python
sns.scatterplot(df['Assault'], df['Cluster2'], hue = df['State'])
plt.legend(bbox_to_anchor = (1,1))
```

# DBSCAN

## Model Building

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps = 1,min_samples = 4)
```

```python
db.fit(x)
```

```python
db.labels_
```

```python
df['Cluster3'] = db.labels_
df.head()
```

```python
df.Cluster3.value_counts()
```

## Visualizing the clusters

```python
df[df['Cluster3']==1]
```

```python
sns.scatterplot(df['Assault'], df['Cluster3'], hue = df['State'])
plt.legend(bbox_to_anchor = (1,1))
```

#### Conclusion:
1. After performing the Clustering of the Airlines data, we find the models that have better clustering ability is DBSCAN > KMeans > Hierarchical.
2. The no. of clusters for for the models are - Hierarchical : 4, KMeans : 4, DBSCAN : 3
